```python
import sys
sys.path.append('scripts')
from importlib import reload
import fonction

df = fonction.load_match('data/ligue1/3900565.json')

print(df)
```

```python
player_name = 'Pape Abdou Paye'

colonnes_joueurs = [col for col in df.columns if 'player' in col.lower()]
print(colonnes_joueurs)
print(df['player.name'].dropna().unique())

nbre_passes = df[df['type.name'] == 'Pass']
print(len(nbre_passes))

player_passes = nbre_passes[nbre_passes['player.name'] == player_name]
print(len(player_passes))

```

```python
essai = [col for col in df.columns if 'pass' in col.lower()]
print(essai)
passereussi = [t for t in df['pass.outcome.name'].unique()]
print(passereussi)

passes_reussies = nbre_passes[nbre_passes['pass.outcome.name'].isna()]
print(len(passes_reussies))
```

```python
player_passes_reussies = passes_reussies[passes_reussies['player.name'] == player_name]

print(len(player_passes_reussies))
```
